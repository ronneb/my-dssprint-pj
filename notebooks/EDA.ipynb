```python
import pandas as pd
import plotly.express as px

# Leer los datos
car_data = pd.read_csv('../vehicles_us.csv')

# Prueba de Histograma (por ejemplo, distribución del kilometraje)
fig_hist = px.histogram(car_data, x="odometer", title="Distribución del Odómetro")
fig_hist.show()

# Prueba de Gráfico de Dispersión (por ejemplo, kilometraje vs precio)
fig_scatter = px.scatter(car_data, x="odometer", y="price", title="Relación entre Odómetro y Precio")
fig_scatter.show()
```
